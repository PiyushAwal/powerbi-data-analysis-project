```python
# Z_test problem solution
import numpy as np
from scipy.stats import norm

sample = [172,174,168,169,171,173,175,170,169,172]
population_mean=170
population_std= 3
sample_mean= np.mean(sample)
n= len(sample)

z_score= (sample_mean - population_mean) / (population_std / np.sqrt(n))
print(z_score)
```

```python
# for calculating the p_value you want to do following things 
p_values= 2*(1-norm.cdf(abs(z_score)))
print(p_values)
```

```python
a = 0.05
if p_values < a:
    print(" i will reject the null hypothesis")
else:
    print("i will accept the null hypothesis")
```
