```python
from itertools import combinations
count =0
list1= ['apple', 'banana', 'mango', 'orange']
com= combinations(list1,2)
for i in com:
    print(i)
    count+= 1
print('combination: ', count)
```

```python
from itertools import permutations
count =0
list1= ['apple', 'banana', 'mango', 'orange']
com= permutations(list1,2)
for i in com:
    print(i)
    count+= 1
print('combination: ', count)
```
